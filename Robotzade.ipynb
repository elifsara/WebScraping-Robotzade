```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

#Selenium kurulumu
def get_selenium_html(url):
    options = Options()
    options.add_argument("--headless")  # Tarayıcıyı arka planda çalıştır
    options.add_argument("--disable-gpu")
    options.add_argument("--window-size=1920,1080")
    
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    driver.get(url)
    
    time.sleep(2)  # Sayfa içeriğinin yüklenmesini bekle

    html = driver.page_source
    driver.quit()
    return html
```

```python
def scrape_robotzade(keyword, sayfa_sayisi=20):
    from selenium import webdriver
    from selenium.webdriver.chrome.service import Service
    from selenium.webdriver.common.by import By
    from selenium.webdriver.chrome.options import Options
    from webdriver_manager.chrome import ChromeDriverManager
    from bs4 import BeautifulSoup
    import pandas as pd
    import time

    # --- Helper: Selenium'dan sayfa içeriğini al ---
    def get_selenium_html(url):
        options = Options()
        options.add_argument("--headless")
        options.add_argument("--disable-gpu")
        options.add_argument("--window-size=1920,1080")
        driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
        driver.get(url)

        time.sleep(2)
        html = driver.page_source
        driver.quit()
        return html

    # --- Sayfa URL'lerini oluştur ---
    pages = [f"https://www.robotzade.com/arama/{keyword}?page={i}" for i in range(1, sayfa_sayisi + 1)]

    veriler = []

    for page_url in pages:
        html = get_selenium_html(page_url)
        soup = BeautifulSoup(html, "html.parser")

        for urun_div in soup.find_all("div", {"class": "col-6 col-lg-4"}):
            ad_kismi = urun_div.find("div", {"class": "showcase-content"})
            showcase_cart = urun_div.find("div", {"class": "showcase-cart"})
            fiyat_kismi = urun_div.find("div", {"class": "showcase-price"})
            fiyatlar = list(fiyat_kismi.stripped_strings) if fiyat_kismi else []
            fiyat = fiyatlar[-1] if len(fiyatlar) > 1 else (fiyatlar[0] if fiyatlar else None)

            ad = ad_kismi.text.strip() if ad_kismi else None
            stok = 1 if showcase_cart and showcase_cart.find("a", {"class": "add-to-cart-button"}) else 0

            veriler.append([ad, fiyat, stok])

    df = pd.DataFrame(veriler, columns=["Ürün Adı", "Fiyat", "Stok"])
    return df
```

```python
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen
```

```python
def getURL(url):
    result = requests.get(url, headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36"})
    soup = BeautifulSoup(result.text, 'html.parser')
    return soup
```

```python
def get_product_links(keyword):
    search_url = "https://www.robotzade.com/arama/{}".format(keyword)
    soup = getURL(search_url)

    links = []

    # Ürün kartlarını bul
    for a_tag in soup.find_all("a", class_="showcase-label-container"):
        href = a_tag.get("href")
        if href:
            full_link = "https://www.robotzade.com" + href
            links.append(full_link)

    return links
```

```python
produks_links = get_product_links("arduino")
```

```python
produks_links
```

```python
def scrape_product_detail(url):
    soup = getURL(url)

    # Ürün adı
    ad_div = soup.find("div", class_="product-title")
    ad = ad_div.text.strip() if ad_div else None

    # Fiyat (indirimli varsa onu al)
    fiyat_divler = soup.find_all("div", class_="product-list-row")
    fiyat_div = fiyat_divler[-1] if fiyat_divler else None


    # Stok kontrolü (stokta yoksa 'Stokta Yok' yazısı var)
    stok_yok_div = soup.find(string=lambda x: x and "Stokta Yok" in x)
    stok = 0 if stok_yok_div else 1

    return {
        "Ürün Adı": ad,
        "Fiyat": fiyat,
        "Stok": stok,
        "Link": url
    }
```

```python
urun_listesi = []
for product_url in produks_links:
    urun_listesi.append(scrape_product_detail(product_url))
```
